6. Find the outliers and replace them outliers

```python
for i in df.columns[2:6]:
  q1 = df[i].quantile(0.25) #Q1
  q3 = df[i].quantile(0.75) #Q3
  print("\nQ1",i,":",q1)
  print("Q3",i,":",q3)
  IQR=q3-q1
  print("IQR",i,":",IQR)
  upper_limit = q3+1.5*IQR
  print("Upper limit",i,":",upper_limit)
  lower_limit =q1-1.5*IQR
  print("Lower limit",i,":",lower_limit)

```

```python
df.median()
```

```python
for i in df.columns[2:5]:
  df[i] = np.where(df[i]>upper_limit,30,df[i])
```
